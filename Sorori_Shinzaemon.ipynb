```
import matplotlib.pyplot as plt
rice_at_first = 1
days = 100
total_grains = [rice_at_first]
daily_grains = [rice_at_first]
for i in range(1,days):
  daily_grains.append(daily_grains[-1]*2)
  total_grains.append(daily_grains[i] + total_grains[-1])
print("Total grains we can receive on 100th day: {}".format(total_grains[-1]))
print("Number of grains we can receive on 100th day: {}".format(daily_grains[-1]))
plt.title('Sosori Shinzaemon problem')
plt.xlabel('days')
plt.ylabel('# of rice')
plt.plot(range(1,days+1),daily_grains, label='daily grains', ls='--')
plt.plot(range(1,days+1),total_grains, label='total grains')
plt.legend()
plt.show()
```

```
def compute_sorori_shinzaemon(initial_num_of_rice = 1,day=100):
    """
    A function that verifies the anecdote of rice by Sorori Shinzaemon. Calculate the number of rice grains per day.
    Parameteres
    --------------
    day: int
        Number of days to get rice (default : 100)
    initial_num_of_rice: int
        Initial number of rice (default: 1)
    Returns
    --------------
    daily_grains : list
        List of the number of rice you get one day (the length of the list is equal to day)
    total_grains : list
        List of the total number of rice we will receive by a certain day (the length of the list will be equal to day)
    """
    total_grains = [initial_num_of_rice]
    daily_grains = [initial_num_of_rice]
    for i in range(1,day):
      daily_grains.append(daily_grains[-1]*2)
      total_grains.append(daily_grains[i] + total_grains[-1])
    return daily_grains, total_grains

# Testing part
day = 10
initial_rice = 2
list_n_grains, list_total_grains = compute_sorori_shinzaemon(initial_rice,day)
plt.title('Sosori Shinzaemon problem')
plt.xlabel('days')
plt.ylabel('# of rice')
plt.plot(range(1,day+1),list_n_grains, label='daily grains', ls='--')
plt.plot(range(1,day+1),list_total_grains, label='total grains')
plt.legend()
plt.show()
```

```
def num_of_people(num_of_days, num_of_grains, calories_per_grain, calories_needed_per_day):
  """
    This function is used to calculate how many people can live within a 
    specific number of days and number of rice grains.
    ----------
    Parameter:
      num_of_days:
        number of days this group of people will survive
      num_of_grains:
        number of total rice grains
      calories_per_grain:
        how many calories per grain
      calories_needed_per_day:
        the amount of calories needed for a person to live for one day
    ----------
    Output:
      The number of people will survive
  """
  total_calories = calories_per_grain * num_of_grains
  return total_calories // (calories_needed_per_day * num_of_days)

# According to https://www.traditionaloven.com/culinary-arts/flours/rice-flour/convert-100-grams-per-100g-to-grain.html#:~:text=How%20many%20grains%20of%20white,same%20white%20rice%20flour%20type.
# There are approximately 1543.24 grains in 100 grams of rice
# Google search result points out 130kcal in 100 grams of rice
grains_per_100gr = 1543.24
calories_per_100gr = 130000
ncalories = calories_per_100gr // grains_per_100gr
# According to https://www.nhs.uk/common-health-questions/food-and-diet/what-should-my-daily-intake-of-calories-be/
# A daily intake of calories per day is 2500 and 2000 for a man and a woman respectively
# On average, I choose the value of 2250 for the total needed amount of calories per day:
calories_needed_per_day = 2250
# Calculate how many people will live within 500 days with the total amount of rice got from sosori shinzaemon at day 100th
list_n_grains = compute_sorori_shinzaemon()[1]
surviving_people = num_of_people(500,list_n_grains[-1], ncalories, calories_needed_per_day)
print("There are {} people can live within 500 days with {} grains of rice".format(surviving_people,list_n_grains[-1]))
```
