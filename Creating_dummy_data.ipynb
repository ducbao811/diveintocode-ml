# Problem 1

```python
import numpy as np
np.random.seed(0)
mean = [-3,0]
covariance = [[1.0,0.8],[0.8,1.0]]
arr = np.random.multivariate_normal(mean,covariance,(500))
print(arr.shape)
```

# Problem 2

```python
import matplotlib.pyplot as plt
plt.scatter(x=arr[:,0],y=arr[:,1], )
```

# Problem 3
In this problem, I use set_xlim for aligning horizontal axis in two histogram

```python
fig, a = plt.subplots(1,2)
x_label = ['x1','x2']
for idx, ax in enumerate(a):
    ax.hist(arr[:,idx])
    ax.set_xlabel(x_label[idx])
    ax.set_ylabel('frequency')
    ax.set_xlim(-6,6)
plt.tight_layout()
```

# Problem 4

```python
mean_prime = [0,-3]
arr_prime = np.random.multivariate_normal(mean_prime,covariance,(500))
plt.scatter(x=arr_prime[:,0],y=arr_prime[:,1], label='1' )
plt.scatter(x=arr[:,0],y=arr[:,1], label='0')
plt.legend()
plt.show()
```

# Problem 5

```python
combination = np.concatenate((arr, arr_prime), axis=0)
print(combination.shape)
```

```python
data = np.concatenate((arr, np.zeros([500,1])), axis=1)
data1 = np.concatenate((arr_prime, np.zeros([500,1])+1), axis=1)
labeling_array = np.concatenate((data,data1), axis=0)
print(labeling_array.shape)
```
